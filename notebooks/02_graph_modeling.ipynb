```python
import pandas as pd

# Load cleaned synthetic dataset
df = pd.read_csv("../data/processed/edges_clean.csv")

# Build adjacency list representation of graph
graph = {}

for _, row in df.iterrows():
    graph.setdefault(row["from_node"], {})[row["to_node"]] = row["cost"]

graph
```

```python
print("Number of nodes:", len(graph))
print("Number of edges:", len(df))
```

```python
graph_clean = {
    int(u): {int(v): int(w) for v, w in nbrs.items()}
    for u, nbrs in graph.items()
}

graph = graph_clean
graph
```
