```python
import pandas as pd
import numpy as np

np.random.seed(42)

nodes = list(range(1, 9))  # 8 delivery locations
edges = []

for i in nodes:
    for j in nodes:
        if i != j and np.random.rand() < 0.4:
            cost = np.random.randint(1, 20)
            edges.append([i, j, cost])

df_edges = pd.DataFrame(edges, columns=["from_node", "to_node", "cost"])

# Save raw dataset
df_edges.to_csv("../data/raw/edges.csv", index=False)

# Clean dataset
df_edges = df_edges.drop_duplicates()
assert (df_edges["cost"] > 0).all()

# Save processed dataset
df_edges.to_csv("../data/processed/edges_clean.csv", index=False)

df_edges
```
